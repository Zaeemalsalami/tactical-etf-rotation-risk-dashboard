```python
# ============================================================
# CELL 2: IMPORT LIBRARIES
# ============================================================

import os
import warnings
from dataclasses import dataclass
from typing import Dict, List, Tuple, Optional

import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px

from scipy import stats
from IPython.display import display

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 120)
pd.set_option("display.float_format", lambda x: f"{x:,.4f}")

print("Libraries imported successfully.")
```

```python
# ============================================================
# CELL 3: PROJECT CONFIGURATION
# ============================================================

@dataclass
class StrategyConfig:
    """
    Configuration object for the tactical ETF rotation strategy.
    Adjust these values to test different strategy assumptions.
    """
    tickers: List[str]
    cash_ticker: str
    equity_benchmark: str
    bond_benchmark: str
    start_date: str
    end_date: Optional[str]
    initial_capital: float
    top_n: int
    transaction_cost_bps: float
    short_window_months: int
    medium_window_months: int
    long_window_months: int
    volatility_window_days: int
    trend_window_days: int
    rolling_metric_window_months: int


CONFIG = StrategyConfig(
    tickers=["SPY", "QQQ", "TLT", "AGG", "GLD", "EEM", "BIL"],
    cash_ticker="BIL",
    equity_benchmark="SPY",
    bond_benchmark="AGG",
    start_date="2010-01-01",
    end_date=None,
    initial_capital=100_000,
    top_n=3,
    transaction_cost_bps=5.0,
    short_window_months=3,
    medium_window_months=6,
    long_window_months=12,
    volatility_window_days=63,
    trend_window_days=200,
    rolling_metric_window_months=12
)

OUTPUT_DIRS = ["data", "charts"]

for folder in OUTPUT_DIRS:
    os.makedirs(folder, exist_ok=True)

print("Configuration loaded.")
print(CONFIG)
```

```python
# ============================================================
# CELL 4: UTILITY FUNCTIONS
# ============================================================

def validate_config(config: StrategyConfig) -> None:
    """
    Validates the strategy configuration before running the project.
    Raises clear errors if important settings are invalid.
    """
    if not config.tickers:
        raise ValueError("Ticker list cannot be empty.")

    if config.cash_ticker not in config.tickers:
        raise ValueError("cash_ticker must be included in tickers.")

    if config.equity_benchmark not in config.tickers:
        raise ValueError("equity_benchmark must be included in tickers.")

    if config.bond_benchmark not in config.tickers:
        raise ValueError("bond_benchmark must be included in tickers.")

    if config.top_n < 1:
        raise ValueError("top_n must be at least 1.")

    if config.initial_capital <= 0:
        raise ValueError("initial_capital must be positive.")

    if config.transaction_cost_bps < 0:
        raise ValueError("transaction_cost_bps cannot be negative.")

    if config.trend_window_days < 20:
        raise ValueError("trend_window_days is too short for a meaningful trend filter.")

    print("Configuration validation passed.")


def safe_month_end_resample(data: pd.DataFrame, how: str = "last") -> pd.DataFrame:
    """
    Converts daily time series into month-end time series.
    Uses 'ME' where supported; falls back to 'M' for older pandas versions.
    """
    if not isinstance(data.index, pd.DatetimeIndex):
        raise TypeError("Data index must be a DatetimeIndex for resampling.")

    try:
        rule = "ME"
        if how == "last":
            return data.resample(rule).last()
        if how == "sum":
            return data.resample(rule).sum()
        if how == "mean":
            return data.resample(rule).mean()
    except Exception:
        rule = "M"
        if how == "last":
            return data.resample(rule).last()
        if how == "sum":
            return data.resample(rule).sum()
        if how == "mean":
            return data.resample(rule).mean()

    raise ValueError("Unsupported resampling method. Use 'last', 'sum', or 'mean'.")


def format_percent(x: float) -> str:
    """
    Formats decimal values as percentages.
    """
    if pd.isna(x):
        return "N/A"
    return f"{x * 100:,.2f}%"


def save_dataframe(df: pd.DataFrame, path: str) -> None:
    """
    Saves a DataFrame to CSV with basic error handling.
    """
    try:
        df.to_csv(path)
        print(f"Saved: {path}")
    except Exception as error:
        print(f"Could not save {path}. Error: {error}")


validate_config(CONFIG)
```

```python
# ============================================================
# CELL 5: DATA COLLECTION PIPELINE
# ============================================================

def download_price_data(config: StrategyConfig) -> pd.DataFrame:
    """
    Downloads adjusted closing prices from Yahoo Finance using yfinance.

    yfinance with auto_adjust=True returns adjusted OHLC prices.
    Therefore, this project uses the 'Close' column as the adjusted close.
    """
    print("Downloading ETF price data...")

    try:
        raw = yf.download(
            tickers=config.tickers,
            start=config.start_date,
            end=config.end_date,
            auto_adjust=True,
            progress=False,
            group_by="column",
            threads=True
        )
    except Exception as error:
        raise RuntimeError(f"Data download failed. Check internet connection or tickers. Error: {error}")

    if raw is None or raw.empty:
        raise ValueError("No data returned from yfinance. Check ticker symbols and date range.")

    if isinstance(raw.columns, pd.MultiIndex):
        if "Close" not in raw.columns.get_level_values(0):
            raise ValueError("Could not find 'Close' prices in downloaded data.")
        prices = raw["Close"].copy()
    else:
        if "Close" not in raw.columns:
            raise ValueError("Could not find 'Close' prices in downloaded data.")
        prices = raw[["Close"]].copy()
        prices.columns = config.tickers[:1]

    prices.index = pd.to_datetime(prices.index)
    prices = prices.sort_index()

    # Remove tickers with no data.
    valid_columns = [col for col in prices.columns if prices[col].dropna().shape[0] > 0]
    missing_columns = sorted(set(config.tickers) - set(valid_columns))

    if missing_columns:
        print(f"Warning: These tickers have no usable data and will be removed: {missing_columns}")

    prices = prices[valid_columns]

    required = [config.cash_ticker, config.equity_benchmark, config.bond_benchmark]
    missing_required = [ticker for ticker in required if ticker not in prices.columns]

    if missing_required:
        raise ValueError(f"Required tickers missing after download: {missing_required}")

    # Forward-fill small gaps and drop rows where all values are missing.
    prices = prices.ffill().dropna(how="all")

    # Keep dates where at least the benchmark has data.
    prices = prices.loc[prices[config.equity_benchmark].notna()]

    if prices.shape[0] < 300:
        raise ValueError("Not enough data for a reliable backtest.")

    print(f"Downloaded {prices.shape[0]} daily observations for {prices.shape[1]} ETFs.")
    return prices


prices = download_price_data(CONFIG)

display(prices.tail())

save_dataframe(prices, "data/prices.csv")
```

```python
# ============================================================
# CELL 6: DATA CLEANING AND RETURN CALCULATION
# ============================================================

def calculate_returns(prices: pd.DataFrame) -> Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    """
    Calculates daily prices, daily returns, and monthly returns.
    """
    if prices.empty:
        raise ValueError("Price data is empty.")

    clean_prices = prices.copy()

    # Remove duplicate dates.
    clean_prices = clean_prices[~clean_prices.index.duplicated(keep="last")]

    # Forward-fill missing values, then drop columns still containing too much missing data.
    clean_prices = clean_prices.ffill()

    missing_ratio = clean_prices.isna().mean()
    usable_cols = missing_ratio[missing_ratio < 0.05].index.tolist()
    removed_cols = sorted(set(clean_prices.columns) - set(usable_cols))

    if removed_cols:
        print(f"Removed tickers with excessive missing values: {removed_cols}")

    clean_prices = clean_prices[usable_cols].dropna()

    daily_returns = clean_prices.pct_change().replace([np.inf, -np.inf], np.nan).dropna(how="all")

    monthly_prices = safe_month_end_resample(clean_prices, how="last")
    monthly_returns = monthly_prices.pct_change().replace([np.inf, -np.inf], np.nan).dropna(how="all")

    print("Return calculation completed.")
    return clean_prices, daily_returns, monthly_returns


clean_prices, daily_returns, monthly_returns = calculate_returns(prices)

display(monthly_returns.tail())

save_dataframe(daily_returns, "data/daily_returns.csv")
save_dataframe(monthly_returns, "data/monthly_returns.csv")
```

```python
# ============================================================
# CELL 7: FEATURE ENGINEERING
# ============================================================

def calculate_momentum_features(
    monthly_prices: pd.DataFrame,
    config: StrategyConfig
) -> Dict[str, pd.DataFrame]:
    """
    Calculates 3-month, 6-month, and 12-month momentum features.
    """
    mom_3m = monthly_prices.pct_change(config.short_window_months)
    mom_6m = monthly_prices.pct_change(config.medium_window_months)
    mom_12m = monthly_prices.pct_change(config.long_window_months)

    weighted_momentum = (
        0.30 * mom_3m +
        0.30 * mom_6m +
        0.40 * mom_12m
    )

    return {
        "mom_3m": mom_3m,
        "mom_6m": mom_6m,
        "mom_12m": mom_12m,
        "weighted_momentum": weighted_momentum
    }


def calculate_trend_filter(
    prices: pd.DataFrame,
    config: StrategyConfig
) -> pd.DataFrame:
    """
    Calculates the 200-day trend filter.
    Signal is True when price is above its moving average.
    """
    moving_average = prices.rolling(config.trend_window_days).mean()
    trend_signal_daily = prices > moving_average
    trend_signal_monthly = safe_month_end_resample(trend_signal_daily.astype(float), how="last").astype(bool)

    return trend_signal_monthly


def calculate_volatility_feature(
    daily_returns: pd.DataFrame,
    config: StrategyConfig
) -> pd.DataFrame:
    """
    Calculates rolling annualized volatility and converts it to month-end frequency.
    """
    daily_volatility = daily_returns.rolling(config.volatility_window_days).std() * np.sqrt(252)
    monthly_volatility = safe_month_end_resample(daily_volatility, how="last")

    return monthly_volatility


def build_features(
    prices: pd.DataFrame,
    daily_returns: pd.DataFrame,
    config: StrategyConfig
) -> Dict[str, pd.DataFrame]:
    """
    Builds all trading features needed for the rotation model.
    """
    monthly_prices = safe_month_end_resample(prices, how="last")

    momentum = calculate_momentum_features(monthly_prices, config)
    trend = calculate_trend_filter(prices, config)
    volatility = calculate_volatility_feature(daily_returns, config)

    # Align all features to the same monthly index.
    common_index = monthly_prices.index
    for feature in momentum.values():
        common_index = common_index.intersection(feature.index)

    common_index = common_index.intersection(trend.index).intersection(volatility.index)

    weighted_momentum = momentum["weighted_momentum"].reindex(common_index)
    trend = trend.reindex(common_index)
    volatility = volatility.reindex(common_index)

    # Avoid division by zero.
    volatility = volatility.replace(0, np.nan)

    risk_adjusted_score = weighted_momentum / volatility
    risk_adjusted_score = risk_adjusted_score.replace([np.inf, -np.inf], np.nan)

    features = {
        "monthly_prices": monthly_prices.reindex(common_index),
        "mom_3m": momentum["mom_3m"].reindex(common_index),
        "mom_6m": momentum["mom_6m"].reindex(common_index),
        "mom_12m": momentum["mom_12m"].reindex(common_index),
        "weighted_momentum": weighted_momentum,
        "trend": trend,
        "volatility": volatility,
        "risk_adjusted_score": risk_adjusted_score
    }

    print("Feature engineering completed.")
    return features


features = build_features(clean_prices, daily_returns, CONFIG)

display(features["risk_adjusted_score"].tail())

save_dataframe(features["risk_adjusted_score"], "data/risk_adjusted_scores.csv")
```

```python
# ============================================================
# CELL 8: SIGNAL GENERATION AND TARGET WEIGHTS
# ============================================================

def build_target_weights(
    features: Dict[str, pd.DataFrame],
    config: StrategyConfig
) -> pd.DataFrame:
    """
    Builds monthly target portfolio weights.

    Logic:
    - Exclude the cash ETF from the risk-asset ranking.
    - Select top N ETFs with positive risk-adjusted momentum.
    - ETF must also pass the trend filter.
    - If no ETF qualifies, allocate 100% to cash.
    """
    scores = features["risk_adjusted_score"].copy()
    trend = features["trend"].copy()

    all_tickers = scores.columns.tolist()
    risk_assets = [ticker for ticker in all_tickers if ticker != config.cash_ticker]

    target_weights = pd.DataFrame(0.0, index=scores.index, columns=all_tickers)

    for date in scores.index:
        date_scores = scores.loc[date, risk_assets].dropna()
        date_trend = trend.loc[date, risk_assets].reindex(date_scores.index).fillna(False)

        eligible_scores = date_scores[(date_scores > 0) & (date_trend == True)]

        if eligible_scores.empty:
            target_weights.loc[date, config.cash_ticker] = 1.0
        else:
            selected = eligible_scores.sort_values(ascending=False).head(config.top_n).index.tolist()
            weight = 1.0 / len(selected)

            for ticker in selected:
                target_weights.loc[date, ticker] = weight

    # Remove dates before enough signals exist.
    target_weights = target_weights.dropna(how="all")

    # Sanity check: weights should sum to 1.
    weight_sums = target_weights.sum(axis=1)
    bad_dates = weight_sums[np.abs(weight_sums - 1.0) > 1e-6]

    if not bad_dates.empty:
        raise ValueError("Some portfolio weights do not sum to 1. Check signal logic.")

    print("Target weights generated.")
    return target_weights


target_weights = build_target_weights(features, CONFIG)

display(target_weights.tail())

save_dataframe(target_weights, "data/target_weights.csv")
```

```python
# ============================================================
# CELL 9: BACKTEST ENGINE
# ============================================================

def calculate_turnover(target_weights: pd.DataFrame) -> pd.Series:
    """
    Calculates portfolio turnover based on changes in target weights.
    Turnover is the absolute change in weights across all positions.
    """
    turnover = target_weights.diff().abs().sum(axis=1)
    turnover.iloc[0] = target_weights.iloc[0].abs().sum()
    return turnover


def run_backtest(
    monthly_returns: pd.DataFrame,
    target_weights: pd.DataFrame,
    config: StrategyConfig
) -> Dict[str, pd.DataFrame]:
    """
    Runs a no-lookahead monthly rebalancing backtest.

    Important:
    - Signals are created at month-end.
    - Weights are shifted by one month.
    - The shifted weights are applied to the next month's returns.
    """
    common_index = monthly_returns.index.intersection(target_weights.index)
    common_columns = monthly_returns.columns.intersection(target_weights.columns)

    rets = monthly_returns.reindex(index=common_index, columns=common_columns).copy()
    weights = target_weights.reindex(index=common_index, columns=common_columns).copy()

    # Shift weights by one month to avoid lookahead bias.
    active_weights = weights.shift(1)

    # Transaction cost from previous rebalance applied to current month.
    turnover = calculate_turnover(weights).shift(1).fillna(0.0)
    cost_decimal = config.transaction_cost_bps / 10_000

    gross_strategy_returns = (active_weights * rets).sum(axis=1)
    transaction_costs = turnover * cost_decimal
    net_strategy_returns = gross_strategy_returns - transaction_costs

    # Remove the first row because weights are unavailable after shifting.
    valid = active_weights.notna().any(axis=1)
    net_strategy_returns = net_strategy_returns.loc[valid]
    gross_strategy_returns = gross_strategy_returns.loc[valid]
    transaction_costs = transaction_costs.loc[valid]
    turnover = turnover.loc[valid]
    active_weights = active_weights.loc[valid]
    rets = rets.loc[valid]

    strategy_value = config.initial_capital * (1 + net_strategy_returns).cumprod()

    results = pd.DataFrame({
        "Gross Strategy Return": gross_strategy_returns,
        "Transaction Cost": transaction_costs,
        "Net Strategy Return": net_strategy_returns,
        "Turnover": turnover,
        "Strategy Value": strategy_value
    })

    print("Backtest completed.")
    return {
        "results": results,
        "active_weights": active_weights,
        "monthly_returns": rets
    }


backtest = run_backtest(monthly_returns, target_weights, CONFIG)

strategy_results = backtest["results"]
active_weights = backtest["active_weights"]
aligned_monthly_returns = backtest["monthly_returns"]

display(strategy_results.tail())

save_dataframe(strategy_results, "data/strategy_results.csv")
save_dataframe(active_weights, "data/active_weights.csv")
```

```python
# ============================================================
# CELL 10: BENCHMARK CONSTRUCTION
# ============================================================

def build_benchmarks(
    monthly_returns: pd.DataFrame,
    strategy_index: pd.DatetimeIndex,
    config: StrategyConfig
) -> pd.DataFrame:
    """
    Builds benchmark return series:
    - SPY buy-and-hold
    - Monthly rebalanced 60/40 portfolio using SPY and AGG
    """
    benchmark_returns = pd.DataFrame(index=strategy_index)

    if config.equity_benchmark not in monthly_returns.columns:
        raise ValueError(f"Missing equity benchmark: {config.equity_benchmark}")

    if config.bond_benchmark not in monthly_returns.columns:
        raise ValueError(f"Missing bond benchmark: {config.bond_benchmark}")

    benchmark_returns["SPY Buy & Hold"] = monthly_returns[config.equity_benchmark].reindex(strategy_index)

    benchmark_returns["60/40 Portfolio"] = (
        0.60 * monthly_returns[config.equity_benchmark].reindex(strategy_index) +
        0.40 * monthly_returns[config.bond_benchmark].reindex(strategy_index)
    )

    benchmark_returns = benchmark_returns.dropna()

    print("Benchmarks created.")
    return benchmark_returns


benchmarks = build_benchmarks(monthly_returns, strategy_results.index, CONFIG)

all_returns = pd.concat(
    [
        strategy_results["Net Strategy Return"].rename("ETF Rotation Strategy"),
        benchmarks
    ],
    axis=1
).dropna()

cumulative_returns = (1 + all_returns).cumprod()

display(all_returns.tail())

save_dataframe(all_returns, "data/all_returns.csv")
```

```python
# ============================================================
# CELL 11: PERFORMANCE METRICS
# ============================================================

def calculate_drawdown(returns: pd.Series) -> pd.Series:
    """
    Calculates drawdown series from return series.
    """
    wealth_index = (1 + returns.dropna()).cumprod()
    running_max = wealth_index.cummax()
    drawdown = wealth_index / running_max - 1
    return drawdown


def calculate_beta_alpha(
    strategy_returns: pd.Series,
    benchmark_returns: pd.Series
) -> Tuple[float, float]:
    """
    Calculates monthly beta and annualized alpha versus benchmark.
    """
    aligned = pd.concat([strategy_returns, benchmark_returns], axis=1).dropna()
    if aligned.shape[0] < 12:
        return np.nan, np.nan

    y = aligned.iloc[:, 0]
    x = aligned.iloc[:, 1]

    if x.var() == 0:
        return np.nan, np.nan

    beta = y.cov(x) / x.var()
    monthly_alpha = y.mean() - beta * x.mean()
    annualized_alpha = monthly_alpha * 12

    return beta, annualized_alpha


def performance_metrics(
    returns: pd.Series,
    benchmark_returns: Optional[pd.Series] = None,
    turnover: Optional[pd.Series] = None,
    risk_free_rate: float = 0.0
) -> Dict[str, float]:
    """
    Calculates professional performance and risk metrics.
    Returns are assumed to be monthly.
    """
    returns = returns.dropna()

    if returns.empty:
        raise ValueError("Return series is empty.")

    periods_per_year = 12
    n_months = len(returns)
    cumulative = (1 + returns).prod()

    cagr = cumulative ** (periods_per_year / n_months) - 1
    annualized_vol = returns.std() * np.sqrt(periods_per_year)

    monthly_rf = risk_free_rate / periods_per_year
    excess_returns = returns - monthly_rf

    sharpe = np.nan
    if returns.std() != 0:
        sharpe = excess_returns.mean() / returns.std() * np.sqrt(periods_per_year)

    downside_returns = returns[returns < 0]
    sortino = np.nan
    if downside_returns.std() != 0 and len(downside_returns) > 1:
        sortino = excess_returns.mean() / downside_returns.std() * np.sqrt(periods_per_year)

    drawdown = calculate_drawdown(returns)
    max_drawdown = drawdown.min()

    calmar = np.nan
    if max_drawdown != 0:
        calmar = cagr / abs(max_drawdown)

    win_rate = (returns > 0).mean()

    var_95 = returns.quantile(0.05)
    cvar_95 = returns[returns <= var_95].mean()

    beta = np.nan
    alpha = np.nan

    if benchmark_returns is not None:
        beta, alpha = calculate_beta_alpha(returns, benchmark_returns)

    avg_monthly_turnover = np.nan
    annualized_turnover = np.nan

    if turnover is not None:
        aligned_turnover = turnover.reindex(returns.index).dropna()
        if not aligned_turnover.empty:
            avg_monthly_turnover = aligned_turnover.mean()
            annualized_turnover = avg_monthly_turnover * 12

    final_growth = cumulative

    return {
        "CAGR": cagr,
        "Annualized Volatility": annualized_vol,
        "Sharpe Ratio": sharpe,
        "Sortino Ratio": sortino,
        "Max Drawdown": max_drawdown,
        "Calmar Ratio": calmar,
        "Win Rate": win_rate,
        "Monthly VaR 95%": var_95,
        "Monthly CVaR 95%": cvar_95,
        "Beta vs SPY": beta,
        "Annualized Alpha vs SPY": alpha,
        "Average Monthly Turnover": avg_monthly_turnover,
        "Annualized Turnover": annualized_turnover,
        "Final Growth Multiple": final_growth
    }


def build_metrics_table(
    all_returns: pd.DataFrame,
    strategy_results: pd.DataFrame,
    config: StrategyConfig
) -> pd.DataFrame:
    """
    Builds a metrics table for the strategy and benchmarks.
    """
    metrics = {}

    spy_returns = all_returns["SPY Buy & Hold"]

    for column in all_returns.columns:
        turnover = strategy_results["Turnover"] if column == "ETF Rotation Strategy" else None

        metrics[column] = performance_metrics(
            returns=all_returns[column],
            benchmark_returns=spy_returns,
            turnover=turnover
        )

    metrics_df = pd.DataFrame(metrics).T

    return metrics_df


metrics_df = build_metrics_table(all_returns, strategy_results, CONFIG)

display(metrics_df)

save_dataframe(metrics_df, "data/performance_metrics.csv")
```

```python
# ============================================================
# CELL 12: CURRENT SIGNAL DASHBOARD
# ============================================================

def build_current_signal_dashboard(
    features: Dict[str, pd.DataFrame],
    target_weights: pd.DataFrame,
    config: StrategyConfig
) -> pd.DataFrame:
    """
    Creates a dashboard of the latest signals, scores, and target weights.
    """
    latest_date = features["risk_adjusted_score"].dropna(how="all").index[-1]

    dashboard = pd.DataFrame(index=features["risk_adjusted_score"].columns)

    dashboard["3M Momentum"] = features["mom_3m"].loc[latest_date]
    dashboard["6M Momentum"] = features["mom_6m"].loc[latest_date]
    dashboard["12M Momentum"] = features["mom_12m"].loc[latest_date]
    dashboard["Weighted Momentum"] = features["weighted_momentum"].loc[latest_date]
    dashboard["Annualized Volatility"] = features["volatility"].loc[latest_date]
    dashboard["Risk-Adjusted Score"] = features["risk_adjusted_score"].loc[latest_date]
    dashboard["Above 200D MA"] = features["trend"].loc[latest_date]
    dashboard["Target Weight"] = target_weights.loc[latest_date]

    dashboard = dashboard.sort_values("Risk-Adjusted Score", ascending=False)

    print(f"Latest signal date: {latest_date.date()}")
    return dashboard


current_dashboard = build_current_signal_dashboard(features, target_weights, CONFIG)

display(current_dashboard)

save_dataframe(current_dashboard, "data/current_signal_dashboard.csv")
```

```python
# ============================================================
# CELL 13: PLOT HELPER FUNCTIONS
# ============================================================

def save_current_figure(filename: str) -> None:
    """
    Saves the current matplotlib figure to the charts directory.
    """
    path = os.path.join("charts", filename)
    plt.savefig(path, dpi=150, bbox_inches="tight")
    print(f"Saved chart: {path}")


def plot_cumulative_returns(cumulative_returns: pd.DataFrame) -> None:
    """
    Plots cumulative returns for the strategy and benchmarks.
    """
    plt.figure(figsize=(12, 6))

    for column in cumulative_returns.columns:
        plt.plot(cumulative_returns.index, cumulative_returns[column], label=column, linewidth=2)

    plt.title("Cumulative Returns: ETF Rotation Strategy vs Benchmarks")
    plt.xlabel("Date")
    plt.ylabel("Growth of $1")
    plt.legend()
    plt.grid(True, alpha=0.3)
    save_current_figure("cumulative_returns.png")
    plt.show()


def plot_drawdowns(all_returns: pd.DataFrame) -> None:
    """
    Plots drawdowns for the strategy and benchmarks.
    """
    plt.figure(figsize=(12, 6))

    for column in all_returns.columns:
        dd = calculate_drawdown(all_returns[column])
        plt.plot(dd.index, dd, label=column, linewidth=2)

    plt.title("Drawdown Comparison")
    plt.xlabel("Date")
    plt.ylabel("Drawdown")
    plt.legend()
    plt.grid(True, alpha=0.3)
    save_current_figure("drawdown.png")
    plt.show()


def plot_allocation_over_time(active_weights: pd.DataFrame) -> None:
    """
    Plots portfolio allocation over time as a stacked area chart.
    """
    weights = active_weights.copy()
    weights = weights.loc[weights.sum(axis=1) > 0]

    plt.figure(figsize=(12, 6))
    plt.stackplot(weights.index, [weights[col] for col in weights.columns], labels=weights.columns)

    plt.title("ETF Allocation Over Time")
    plt.xlabel("Date")
    plt.ylabel("Portfolio Weight")
    plt.legend(loc="upper left", ncol=4)
    plt.grid(True, alpha=0.3)
    save_current_figure("allocation_over_time.png")
    plt.show()


def plot_latest_scores(current_dashboard: pd.DataFrame) -> None:
    """
    Plots latest risk-adjusted scores by ETF.
    """
    score_data = current_dashboard["Risk-Adjusted Score"].dropna().sort_values()

    plt.figure(figsize=(10, 6))
    plt.barh(score_data.index, score_data.values)

    plt.title("Latest Risk-Adjusted Momentum Scores")
    plt.xlabel("Score")
    plt.ylabel("ETF")
    plt.grid(True, axis="x", alpha=0.3)
    save_current_figure("latest_scores.png")
    plt.show()


def plot_rolling_sharpe(all_returns: pd.DataFrame, window: int) -> None:
    """
    Plots rolling Sharpe ratio for the strategy and benchmarks.
    """
    rolling_sharpe = (
        all_returns.rolling(window).mean() /
        all_returns.rolling(window).std()
    ) * np.sqrt(12)

    plt.figure(figsize=(12, 6))

    for column in rolling_sharpe.columns:
        plt.plot(rolling_sharpe.index, rolling_sharpe[column], label=column, linewidth=2)

    plt.title(f"Rolling {window}-Month Sharpe Ratio")
    plt.xlabel("Date")
    plt.ylabel("Sharpe Ratio")
    plt.legend()
    plt.grid(True, alpha=0.3)
    save_current_figure("rolling_sharpe.png")
    plt.show()


def plot_rolling_volatility(all_returns: pd.DataFrame, window: int) -> None:
    """
    Plots rolling annualized volatility.
    """
    rolling_vol = all_returns.rolling(window).std() * np.sqrt(12)

    plt.figure(figsize=(12, 6))

    for column in rolling_vol.columns:
        plt.plot(rolling_vol.index, rolling_vol[column], label=column, linewidth=2)

    plt.title(f"Rolling {window}-Month Annualized Volatility")
    plt.xlabel("Date")
    plt.ylabel("Annualized Volatility")
    plt.legend()
    plt.grid(True, alpha=0.3)
    save_current_figure("rolling_volatility.png")
    plt.show()


def plot_correlation_heatmap(monthly_returns: pd.DataFrame) -> None:
    """
    Plots correlation heatmap of ETF monthly returns.
    """
    corr = monthly_returns.corr()

    fig = px.imshow(
        corr,
        text_auto=".2f",
        title="ETF Monthly Return Correlation Heatmap",
        aspect="auto"
    )

    fig.write_html("charts/correlation_heatmap.html")
    fig.show()

    print("Saved interactive chart: charts/correlation_heatmap.html")


def plot_monthly_returns_heatmap(strategy_returns: pd.Series) -> None:
    """
    Plots monthly returns heatmap for the ETF rotation strategy.
    """
    returns = strategy_returns.dropna().copy()
    heatmap_df = pd.DataFrame({
        "Year": returns.index.year,
        "Month": returns.index.month,
        "Return": returns.values
    })

    pivot = heatmap_df.pivot(index="Year", columns="Month", values="Return")
    pivot = pivot.sort_index()

    fig = px.imshow(
        pivot,
        text_auto=".1%",
        title="ETF Rotation Strategy Monthly Returns Heatmap",
        aspect="auto"
    )

    fig.write_html("charts/monthly_returns_heatmap.html")
    fig.show()

    print("Saved interactive chart: charts/monthly_returns_heatmap.html")
```

```python
# ============================================================
# CELL 14: GENERATE PROFESSIONAL VISUALIZATIONS
# ============================================================

plot_cumulative_returns(cumulative_returns)
plot_drawdowns(all_returns)
plot_allocation_over_time(active_weights)
plot_latest_scores(current_dashboard)
plot_rolling_sharpe(all_returns, CONFIG.rolling_metric_window_months)
plot_rolling_volatility(all_returns, CONFIG.rolling_metric_window_months)
plot_correlation_heatmap(aligned_monthly_returns)
plot_monthly_returns_heatmap(all_returns["ETF Rotation Strategy"])
```

```python
# ============================================================
# CELL 15: INTERACTIVE PLOTLY PERFORMANCE CHART
# ============================================================

def interactive_cumulative_returns(cumulative_returns: pd.DataFrame) -> None:
    """
    Creates an interactive Plotly cumulative return chart.
    """
    fig = go.Figure()

    for column in cumulative_returns.columns:
        fig.add_trace(
            go.Scatter(
                x=cumulative_returns.index,
                y=cumulative_returns[column],
                mode="lines",
                name=column
            )
        )

    fig.update_layout(
        title="Interactive Cumulative Returns",
        xaxis_title="Date",
        yaxis_title="Growth of $1",
        hovermode="x unified",
        template="plotly_white"
    )

    fig.write_html("charts/interactive_cumulative_returns.html")
    fig.show()

    print("Saved interactive chart: charts/interactive_cumulative_returns.html")


interactive_cumulative_returns(cumulative_returns)
```
